# Initial Data Distribution for Numerical Columns

In this section, we will visualize the data distribution for numerical columns in the dataset using histograms.

```python
df.hist()
plt.show()
```

# Observations of the Histograms

1. Data Distribution: Understand how data is distributed across different values for each numerical column. You can identify whether the data follows a normal distribution, is skewed to the left or right, or exhibits another distribution pattern.

2. Frequency of Values: Histograms show the frequency of data points within specific ranges, allowing you to see which values are more common.

3. Identify Outliers and Anomalies: Outliers or unusual data points can be detected as isolated bars in the histogram, which might warrant further investigation.

4. Skewness of Data: The direction and extent of skewness (right or left) can be observed, indicating whether high or low values dominate.

5. Data Variability: The spread of the histogram bars indicates the variability or dispersion of data within each column.

6. Comparison Across Variables: If there are multiple numerical columns, histograms for each allow for comparison across different variables, helping identify patterns or inconsistencies.

## Initial Data Distribution for Categorical Columns

In this section, we will explore the distribution of data across categorical columns in the dataset.

Value Counts for Individual Categorical Column

Let's start by inspecting the value counts 

```python
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()
```

# Observations from Value Counts and Count Plots

Distribution of Other Categorical Columns

1. Prevalence of Categories: Identifies which categories are most and least common, giving insight into the distribution's skewness towards certain values.

2. Category Diversity: Reveals the variety of categories present in each column, indicating the complexity and granularity of the data.

3. Imbalance in Categories: Highlights any significant imbalances among category frequencies, which could impact data analysis and model performance.

4. Data Quality and Consistency: May uncover potential issues with data quality, such as inconsistent categorization, which could necessitate data cleaning or preprocessing.

5. Visual Comparison: Allows for an easy visual comparison between the frequencies of different categories within a column, facilitating a quick understanding of the data structure and composition.
